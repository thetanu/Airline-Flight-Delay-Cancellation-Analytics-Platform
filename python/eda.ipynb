# ✈️ Exploratory Data Analysis (EDA) - Flight Delays & Cancellations
This notebook explores the training dataset containing operations, weather profiles, and airline metrics to understand the patterns behind flight delays.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Load Data
We load a sample of 100,000 rows for fast analysis.

```python
train_path = r'../archive (5)/train.csv'
df = pd.read_csv(train_path, nrows=100000)
print(f"Data Shape: {df.shape}")
df.head()
```

## 2. Inspect Target Distribution
The target variable is `DEP_DEL15` representing a departure delay of 15 minutes or more.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='DEP_DEL15', data=df, hue='DEP_DEL15', palette='Blues', legend=False)
plt.title("Distribution of Departure Delays (>15 mins)")
plt.xlabel("Is Delayed (0 = On-Time, 1 = Delayed)")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

print("Class Proportion:")
print(df['DEP_DEL15'].value_counts(normalize=True))
```

## 3. Delay Rates by Carrier
Let's visualize which airlines historically experience the highest rates of departure delays.

```python
plt.figure(figsize=(10, 6))
airline_delays = df.groupby('CARRIER_NAME')['DEP_DEL15'].mean().reset_index().sort_values(by='DEP_DEL15', ascending=False)
sns.barplot(x='DEP_DEL15', y='CARRIER_NAME', data=airline_delays, hue='CARRIER_NAME', palette='viridis', legend=False)
plt.title("Historical Delay Rate by Airline Carrier")
plt.xlabel("Delay Rate (%)")
plt.ylabel("Airline")
plt.tight_layout()
plt.show()
```

## 4. Delay Rates by Time Block
Flights scheduled in the afternoon and evening are subject to cumulative delays due to cascading operations.

```python
plt.figure(figsize=(10, 5))
block_delays = df.groupby('DEP_TIME_BLK')['DEP_DEL15'].mean().reset_index().sort_values(by='DEP_TIME_BLK')
sns.lineplot(x='DEP_TIME_BLK', y='DEP_DEL15', data=block_delays, marker='o', color='#1F4E78', linewidth=2)
plt.xticks(rotation=45)
plt.title("Delay Probability by Scheduled Departure Time Slot")
plt.xlabel("Departure Time Block")
plt.ylabel("Delay Rate (%)")
plt.tight_layout()
plt.show()
```

## 5. Weather & Features Correlation Heatmap
Let's inspect how weather variables (precipitation, snow, wind speeds) relate to delays.

```python
weather_cols = ['DEP_DEL15', 'PRCP', 'SNOW', 'SNWD', 'TMAX', 'AWND', 'PLANE_AGE']
corr_matrix = df[weather_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.3f', linewidths=0.5, vmin=-1.0, vmax=1.0)
plt.title("Correlation Matrix: Delays, Weather, and Aircraft Age")
plt.tight_layout()
plt.show()
```
